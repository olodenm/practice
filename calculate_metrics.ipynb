```python
df = pd.read_csv('/content/new_dataframe.csv')
df
```

```python
def calculate_metrics(df, columns):
    for column in columns:
        mean = df[column].mean()
        median = df[column].median()
        percentile_95 = df[column].quantile(0.95)

        print(f"Среднее значение {column}: {mean}")
        print(f"Медиана {column}: {median}")
        print(f"95-й перцентиль {column}: {percentile_95}")

        # Распределение частот
        plt.figure(figsize=(10, 4))
        plt.subplot(1, 2, 1)
        sns.histplot(df[column], kde=True)
        plt.title(f"Распределение частот {column}")

        # Ящик с усами
        plt.subplot(1, 2, 2)
        sns.boxplot(x=df[column])
        plt.title(f"Ящик с усами {column}")
        plt.show()
```

```python
calculate_metrics(df, ['age', 'hour', 'distance'])
```
